```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
data1 = pd.read_csv("NewspaperData.csv")
data1
```

```python
data1.info()
```

```python
data1[data1.duplicated(keep=False)]
```

```python
cols = data1.columns
colors = ['black','red']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar = True)
```

```python
fig, axes = plt.subplots(2, 1,figsize=(8,6), gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["Newspaper"],ax=axes[0],color='yellow',width=0.5,orient = 'h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Newspaper Levels")
sns.histplot(data1["Newspaper"],kde=True, ax=axes[1],color='blue',bins=30)
axes[1].set_title("Histogram with kde")
axes[1].set_xlabel("Newspaper levels")
axes[1].set_ylabel("frequency")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 1,figsize=(8,6), gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["daily"],ax=axes[0],color='yellow',width=0.5,orient = 'h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Newspaper Levels")
sns.histplot(data1["daily"],kde=True, ax=axes[1],color='blue',bins=30)
axes[1].set_title("Histogram with kde")
axes[1].set_xlabel("Newspaper levels")
axes[1].set_ylabel("frequency")
plt.tight_layout()
plt.show()
```

### Observations
- There are no missing values
- There are no duplicate values
